```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Load the CSV data into a pandas DataFrame
file_path = r'D:\EVchargingdata.csv'
df = pd.read_csv(file_path)

# Function to parse and handle invalid dates
def parse_date(date_str):
    if isinstance(date_str, str):
        try:
            return datetime.strptime(date_str, '%m/%d/%Y %H:%M')
        except ValueError:
            pass
    return None

# Apply the parse_date function to 'Transaction Date (Pacific Time)'
df['Transaction Date (Pacific Time)'] = df['Transaction Date (Pacific Time)'].apply(parse_date)

# Remove rows with invalid dates
df = df.dropna(subset=['Transaction Date (Pacific Time)'])

# Sort the DataFrame by 'Transaction Date (Pacific Time)'
df = df.sort_values(by='Transaction Date (Pacific Time)')

# Create 8 different graphs

# 1. Line plot for Transaction Date vs Charging Time
plt.figure(figsize=(12, 6))
plt.plot(df['Transaction Date (Pacific Time)'], df['Charging Time (hh:mm:ss)'], label='Charging Time')
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('Charging Time (hh:mm:ss)')
plt.title('Transaction Date vs Charging Time')
plt.legend()
plt.grid()

# 2. Scatter plot for Transaction Date vs Charging Time
plt.figure(figsize=(12, 6))
plt.scatter(df['Transaction Date (Pacific Time)'], df['Charging Time (hh:mm:ss)'], label='Charging Time', alpha=0.5)
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('Charging Time (hh:mm:ss)')
plt.title('Transaction Date vs Charging Time (Scatter)')
plt.legend()
plt.grid()

# 3. Line plot for Transaction Date vs Energy
plt.figure(figsize=(12, 6))
plt.plot(df['Transaction Date (Pacific Time)'], df['Energy (kWh)'], label='Energy (kWh)')
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('Energy (kWh)')
plt.title('Transaction Date vs Energy')
plt.legend()
plt.grid()

# 4. Scatter plot for Transaction Date vs Energy
plt.figure(figsize=(12, 6))
plt.scatter(df['Transaction Date (Pacific Time)'], df['Energy (kWh)'], label='Energy (kWh)', alpha=0.5)
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('Energy (kWh)')
plt.title('Transaction Date vs Energy (Scatter)')
plt.legend()
plt.grid()

# 5. Line plot for Transaction Date vs GHG Savings
plt.figure(figsize=(12, 6))
plt.plot(df['Transaction Date (Pacific Time)'], df['GHG Savings (kg)'], label='GHG Savings (kg)')
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('GHG Savings (kg)')
plt.title('Transaction Date vs GHG Savings')
plt.legend()
plt.grid()

# 6. Scatter plot for Transaction Date vs GHG Savings
plt.figure(figsize=(12, 6))
plt.scatter(df['Transaction Date (Pacific Time)'], df['GHG Savings (kg)'], label='GHG Savings (kg)', alpha=0.5)
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('GHG Savings (kg)')
plt.title('Transaction Date vs GHG Savings (Scatter)')
plt.legend()
plt.grid()

# 7. Line plot for Transaction Date vs Gasoline Savings
plt.figure(figsize=(12, 6))
plt.plot(df['Transaction Date (Pacific Time)'], df['Gasoline Savings (gallons)'], label='Gasoline Savings (gallons)')
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('Gasoline Savings (gallons)')
plt.title('Transaction Date vs Gasoline Savings')
plt.legend()
plt.grid()

# 8. Scatter plot for Transaction Date vs Gasoline Savings
plt.figure(figsize=(12, 6))
plt.scatter(df['Transaction Date (Pacific Time)'], df['Gasoline Savings (gallons)'], label='Gasoline Savings (gallons)', alpha=0.5)
plt.xlabel('Transaction Date (Pacific Time)')
plt.ylabel('Gasoline Savings (gallons)')
plt.title('Transaction Date vs Gasoline Savings (Scatter)')
plt.legend()
plt.grid()

# To make Y-axis labels visible for the scatter plots
plt.tick_params(axis='y', which='both', left=True, right=True)

plt.show()
```
